Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.

```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
temp_arr = np.random.uniform(-5, 35, size=(10, 12))
temp_arr = np.round(temp_arr).astype(int)
temp_arr
```

```python
towns=['Acrine', 'Blencogo', 'Cherrytown', 'Kinecardine', 'Saker Keep', 'Queenstown', 'Howe', 'Aelinmiley', 'Aermagh', 'Kelna']
df= pd.DataFrame(temp_arr, index=range(len(towns)), columns=['January', 'February', 'March',
                                                                     'April', 'May', 'June', 'July',
                                                                       'August', 'September', 'October',
                                                                       'November', 'December'])
df
```

Calculate the annual average temperature for each city.

```python
df1 = df
df1['ann_avr'] = round(df.mean(axis = 1)).astype(int)
df1
```

Identify the city with the highest and lowest average temperature for the year.

```python
df1['ann_avr'].max()
```

```python
towns[df1['ann_avr'].idxmax()]
```

```python
df1['ann_avr'].min()
```

```python
towns[df1['ann_avr'].idxmin()]
```

```python
df.reset_index(inplace=True)
df.rename(columns={'index': 'Town'}, inplace=True)


df_melted = df.melt(id_vars='Town', var_name='Month', value_name='Temperature')


fig = px.line(df_melted, x='Month', y='Temperature', color='Town',
              title='Average Monthly Temperature for 10 Towns',
              labels={'Temperature': 'Temperature (°C)', 'Month': 'Month', 'Town': 'Town'})

fig.show()
```
