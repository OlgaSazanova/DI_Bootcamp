```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('imdb_top_1000.csv')
data.info()
```

```python
data.head()
```

```python
data['Released_Year'].unique()
```

```python
data[data['Released_Year']=='PG']
```

```python
data['Released_Year']=data['Released_Year'].replace('PG', '1995')
```

```python
data.duplicated().sum()
```

```python
data['Gross'] = data['Gross'].str.replace(',', '')
data.head()
```

```python
data['Gross'] = data['Gross'].fillna(0)
```

```python
data = data.astype({"Released_Year": int, "Gross": int})
data.info()
```

Create a line plot with Matplotlib showing trends in IMDB_Rating over the years.

```python
import matplotlib.pyplot as plt
```

```python
mean_ratings_by_year = data.groupby('Released_Year')['IMDB_Rating'].mean().reset_index()

# Plot the trend
plt.figure(figsize=(10, 6))
plt.plot(mean_ratings_by_year['Released_Year'], mean_ratings_by_year['IMDB_Rating'], marker='o')
plt.xlabel('Year')
plt.ylabel('Mean IMDB Rating')
plt.title('Trends in IMDB Ratings Over the Years')
plt.grid(True)
plt.show()
```

Generate a Seaborn bar plot to compare the number of movies across different genres.

```python
data['Genre'].unique()
```

```python
genres_df = data['Genre'].str.split(', ', expand=True)
genres_df 
```

```python
genres_unpivoted = genres_df.melt(value_name='Genre').dropna()['Genre']

# Count the occurrences of each genre
genre_counts = genres_unpivoted.value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']

# Create the bar plot using Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Genre', y='Count', data=genre_counts, palette='viridis')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Across Different Genres')
plt.xticks(rotation=45)
plt.show()
```

```python
genre_counts.head()
```

Use a Matplotlib scatter plot to examine the relationship between directors and movie ratings.

```python

top_directors = data['Director'].value_counts().head(10).index

# Filter the dataset to include only the top directors
filtered_data = data[data['Director'].isin(top_directors)]

# Create a scatter plot using Seaborn
plt.figure(figsize=(12, 8))
sns.scatterplot(x='IMDB_Rating', y='Director', data=filtered_data, alpha=0.7)
plt.title('IMDB Rating vs. Top Directors')
plt.xlabel('IMDB Rating')
plt.ylabel('Director')
plt.show()
```

```python
sns.scatterplot(x=data['IMDB_Rating'], y=data['Released_Year'], alpha=0.5)
plt.title('IMDB_Rating vs. Released_Year')
plt.xlabel('IMDB_Rating')
plt.ylabel('Released_Year')
plt.show()
```

Employ a Seaborn pairplot to explore relationships involving Stars and their impact on IMDB_Rating and Gross.

```python
data_filtered = data[['Star1', 'IMDB_Rating', 'Gross']]

# Convert 'Stars' to a category type for proper plotting
data_filtered['Gross'] = data_filtered['Gross'].astype('category')

# Plot pairplot
sns.pairplot(data_filtered, hue='Gross', diag_kind='kde')

# Show plot
plt.show()
```

```python
data_filtered.head()
```

Visualize the distribution of IMDB_Rating across different Genres using a Seaborn box plot.

```python

data['Genre1'] = data['Genre'].str.split(',').str[0]
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre1', y='IMDB_Rating', data=data)

# Add title and labels
plt.title('Distribution of IMDB Ratings Across Different Genres')
plt.xlabel('Genre')
plt.ylabel('IMDB Rating')

# Show the plot
plt.show()
```

Create a heatmap using Seaborn to identify correlations between numerical columns like IMDB_Rating, Meta_score, No_of_votes, and Gross.

```python
correlation_matrix = data[['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross']].corr()


sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```
