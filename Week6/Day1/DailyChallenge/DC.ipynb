```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('2019.csv')
data.head(24)
```

We don't have any missing values, all data types are correct. We don't have any duplicates.

Employ a Matplotlib scatter plot to investigate the relationship between ‘Social support’ and ‘Score’.

```python
x = data['Social support']
y = data['Score']

plt.scatter(x, y)
plt.title("The relationship between ‘Social support’ and ‘Score’")
plt.xlabel("Social support")
plt.ylabel("Score")
plt.show()
```

Create a Matplotlib subplot that compares ‘GDP per Capita’ and ‘Healthy Life Expectancy’ across different regions.

```python
data_CIS = data[data['Country or region'].isin(['Armenia', 'Belarus', 'Russia', 'Ukraine', 'Uzbekistan', 'Tajikistan', 'Kazakhstan', 'Turkmenistan', 'Azerbaijan' ])]
data_CIS
```

```python
fig, axes = plt.subplots(2, 1, figsize = (7, 6), sharex=True)


# Plot GDP per Capita
axes[0].bar(data_CIS['Country or region'], data_CIS['GDP per capita'], color='blue')
axes[0].set_title('GDP per Capita across Countries')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('GDP per Capita')
axes[0].tick_params(axis='x', rotation=90)

# Plot Healthy Life Expectancy
axes[1].bar(data_CIS['Country or region'], data_CIS['Healthy life expectancy'], color='green')
axes[1].set_title('Healthy Life Expectancy across Countries')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Healthy Life Expectancy')
axes[1].tick_params(axis='x', rotation=90)

# Adjust layout
plt.tight_layout()
plt.show()
```

Use a bar plot for ‘GDP per Capita’ and a line plot for ‘Healthy Life Expectancy’ on the same axes to observe how economic strength relates to health outcomes in different regions.

```python
fig, ax1 = plt.subplots(figsize=(10, 7))

# Bar plot for GDP per Capita
ax1.bar(data_CIS['Country or region'], data_CIS['GDP per capita'], color='blue', alpha=0.6, label='GDP per Capita')
ax1.set_xlabel('Country or region')
ax1.set_ylabel('GDP per capita', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.tick_params(axis='x', rotation=90)

# Create a secondary y-axis for Healthy Life Expectancy
ax2 = ax1.twinx()
ax2.plot(data_CIS['Country or region'], data_CIS['Healthy life expectancy'], color='green', marker='o', label='Healthy Life Expectancy')
ax2.set_ylabel('Healthy life expectancy', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Add titles and legends
plt.title('GDP per Capita and Healthy Life Expectancy across Countries')
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```
