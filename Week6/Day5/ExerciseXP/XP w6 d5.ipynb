```python
import pandas as pd
import plotly.express as px
```

```python
data=pd.read_csv('Sample - Superstore.csv', encoding='latin-1')
data.info()
```

```python
data.head()
```

Which states have the most sales?


```python
top_states = data.groupby('State')['Sales'].sum().sort_values(ascending=False).head().reset_index(name='count')
top_states
```

```python
fig = px.bar(top_states, x='State', y='count', color='State', title="Sales by States")
fig.show()
```

What is the difference between New York and California in terms of sales and profit? (Compare the total sales and profit between New York and California.)

```python
data['State'].unique()
```

```python
NY_sales = data[data['State']== 'New York']['Sales'].sum().round().astype(int)
NY_sales 
```

```python
NY_profit = data[data['State']== 'New York']['Profit'].sum().round().astype(int)
NY_profit
```

```python
C_sales = data[data['State']== 'California']['Sales'].sum().round().astype(int)
C_sales 
```

```python
C_profit = data[data['State']== 'California']['Profit'].sum().round().astype(int)
C_profit
```

```python

data1 = {
    'State': ['New York', 'New York', 'California', 'California'],
    'Category': ['Sales', 'Profit', 'Sales', 'Profit'],
    'Amount': [NY_sales, NY_profit, C_sales, C_profit]
}
df = pd.DataFrame(data1)
df
```

```python
fig = px.bar(df, x='State', y='Amount', color='Category', barmode='group',
             title="Compare Sales and Profit for New York and California")
fig.show()
```

We see an interesting thing: Profits are almost the same in two states but Sales on California much more than in New York (458k VS 311k). It means that sales in New York are more profitable.

What are the Top 20 cities by Sales ? What about the Top 20 cities by Profit ? Are there any difference among cities in profitability ? (Identify the top 20 cities based on total sales and total profit and analyze differences in profitability among these cities.)

```python
top_cities_sales = data.groupby('City')['Sales'].sum().sort_values(ascending = False).head(20).reset_index()
top_cities_sales
```

```python
top_cities_profit = data.groupby('City')['Profit'].sum().sort_values(ascending = False).head(20).reset_index()
top_cities_profit
```

```python
top_cities = pd.merge(top_cities_sales, top_cities_profit, on='City')

top_cities
```

```python
fig = px.scatter(top_cities, x='Sales', y='Profit', text='City',
                 title='Sales vs Profit for Top Cities')


fig.update_traces(textposition='top center')


fig.show()
```

What are the Top 20 customers by Sales?

```python
top_customers = data.groupby(['Customer ID', 'Customer Name'])['Sales'].sum().sort_values(ascending = False).head(20).reset_index()
top_customers
```

Plot the Cumulative curve in Sales by Customers. Can we apply Pareto principle to customers and Sales ?

```python
top_customers['Cumulative Sales'] = top_customers['Sales'].cumsum()
top_customers['Cumulative % of Sales'] = top_customers['Cumulative Sales'] / top_customers['Sales'].sum() * 100
top_customers['Cumulative % of Customers'] = (top_customers.index + 1) / len(top_customers) * 100


fig = px.line(top_customers, x='Cumulative % of Customers', y='Cumulative % of Sales',
              title='Cumulative Sales by Customers (Pareto Principle)',
              labels={'Cumulative % of Customers': 'Cumulative % of Customers', 'Cumulative % of Sales': 'Cumulative % of Sales'})


fig.show()
```

No, we can't apply Pareto principle to customers and Sales
