```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_excel('US Superstore data.xls')
```

```python
data.head()
```

```python
data.info()
```

```python
data['order_year']= data['Order Date'].dt.year
data.info()
```

```python
data.head()
```

```python
df2 = data.groupby(['Category', 'order_year'])['Sales'].count().reset_index()
df2
```

```python
fig = px.line(df2, x="order_year", y="Sales", color='Category')
fig.show()
```

We see that most of sales were sales of Ofice Supplies, and these sales increase over the time.

Build an interactive map to visualize sales distribution by country.

```python
data['Country'].nunique()
```

It is just one country, so let's use State

```python
sales_by_state = data.groupby('State')['Sales'].sum().reset_index()

fig = px.choropleth(sales_by_state, 
                    locations="State",
                    locationmode="USA-states",
                    color="Sales",
                    hover_name="State",
                    color_continuous_scale=px.colors.sequential.Plasma,
                    scope="usa",
                    title="Sales Distribution by State")


fig.show()
```

```python
sales_by_state.sort_values(by='Sales', ascending = False).head(10)
```

California and New York are  on the top of sales

```python
top_by_sales = data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10).reset_index(name='Sales')
top_by_sales
```

Create a scatter plot to analyze the relationship between profit and discount.

```python
fig = px.scatter(data, x='Profit', y='Discount',
             color_continuous_scale='Viridis', 
             labels={'Profit': 'Profit', 'Discount': 'Discount'},
             title='The relationship between profit and discount')

# Display the plot
fig.show()
```

We see that profit is greater when discount is less. And when discount greater - profit is less.
