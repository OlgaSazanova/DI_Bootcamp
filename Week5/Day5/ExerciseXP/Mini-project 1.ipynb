We have dataset 'Mental Health in Tech Survey', it has 27 columns. Most of the columns has data object type, because questions of survey are subjective and trying to understand how person handle with his mental problems. We have only one column that quantitative - 'Age'. We have  one more column that could be quantitative - 'no_employees', but it has number range, not just one number.

We have some None valuues in columns:
<br>  state    because it only for people from USA
<br> work_interfere  (question: If you have a mental health condition, do you feel that it interferes with your work?). So it means that people ho didn't answer don't have mental illness
<br> comments - because it was optional

let's fill NA in column work_interfere

```python
data['work_interfere'] = data['work_interfere'].fillna('No illness')
data.info()
```

```python
data['Age'].unique()
```

We see that we have here some  impossible numbers, so let's check what will if we filter them

```python
data[(data['Age'] >= 16) & (data['Age'] <90)].count()
```

Ok,  we have just 8 rows with incorrect age, we can delete it from our dataframe

```python
data = data[(data['Age'] >= 16) & (data['Age'] <90)]
```

```python
data['Age'].value_counts().sort_index(ascending=True)
```

```python
pd.options.mode.chained_assignment = None 
```

```python
bins = [17, 24, 30, 35, 40, 50, 80]
labels = ['18-24', '25-30', '31-35', '36-40', '41-50', '51+']
```

```python
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels, right=True)

data.head()
```

What is the distribution of mental health conditions among different age groups in the tech industry?


```python
data[data['work_interfere'] != 'No illness'].groupby('AgeGroup')['work_interfere'].count()
```

So we see that the most number of people in age 25-30. The less older than 51.

How does the frequency of mental health issues vary by gender?


```python
data[data['work_interfere'] != 'No illness'].groupby('Gender')['work_interfere'].count()
```

Identify the countries with the highest and lowest reported rates of mental health issues in the tech industry.

```python
data[data['work_interfere'] != 'No illness'].groupby('Country')['work_interfere'].count().sort_values(ascending=False)
```
